# 00: Data Loading & Validation

This notebook demonstrates how to load and validate code submissions from CSV and JSON formats.

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.io import load_submissions, validate_submissions, SubmissionValidator

# Set style for better visualizations
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

## 1. Creating Sample Data

First, let's create sample CSV and JSON files for demonstration.

```python
import json

# Sample submissions
sample_data = [
    {
        'submission_id': 's001',
        'language': 'python',
        'code': 'def factorial(n):\n    if n <= 1:\n        return 1\n    return n * factorial(n-1)'
    },
    {
        'submission_id': 's002',
        'language': 'python',
        'code': 'def fact(x):\n    if x <= 1:\n        return 1\n    return x * fact(x-1)'
    },
    {
        'submission_id': 's003',
        'language': 'python',
        'code': 'def factorial(n):\n    result = 1\n    for i in range(2, n+1):\n        result *= i\n    return result'
    },
    {
        'submission_id': 's004',
        'language': 'python',
        'code': 'def compute_factorial(num):\n    from math import factorial\n    return factorial(num)'
    },
    {
        'submission_id': 's005',
        'language': 'python',
        'code': 'def factorial(n):\n    if n == 0 or n == 1:\n        return 1\n    else:\n        return n * factorial(n - 1)'
    }
]

# Save as CSV
df = pd.DataFrame(sample_data)
df.to_csv('../data/raw/sample_submissions.csv', index=False)
print("âœ“ Created sample_submissions.csv")

# Save as JSON
with open('../data/raw/sample_submissions.json', 'w') as f:
    json.dump({'submissions': sample_data}, f, indent=2)
print("âœ“ Created sample_submissions.json")

print(f"\nCreated {len(sample_data)} sample submissions")
```

## 2. Loading from CSV

Load submissions from CSV format.

```python
# Load CSV
submissions_csv = load_submissions('../data/raw/sample_submissions.csv')

print(f"Loaded {len(submissions_csv)} submissions from CSV")
print("\nFirst submission:")
print(f"  ID: {submissions_csv[0]['submission_id']}")
print(f"  Language: {submissions_csv[0]['language']}")
print(f"  Code preview: {submissions_csv[0]['code'][:50]}...")
```

## 3. Loading from JSON

Load submissions from JSON format.

```python
# Load JSON
submissions_json = load_submissions('../data/raw/sample_submissions.json')

print(f"Loaded {len(submissions_json)} submissions from JSON")
print("\nBoth formats loaded successfully!")
print(f"CSV count: {len(submissions_csv)}")
print(f"JSON count: {len(submissions_json)}")
```

## 4. Data Validation

Validate that all submissions meet requirements.

```python
# Validate submissions
is_valid, errors = validate_submissions(submissions_csv, strict=False)

if is_valid:
    print("âœ… All submissions are VALID")
else:
    print("âŒ Validation FAILED")
    for error in errors:
        print(f"  - {error}")
```

## 5. Submission Statistics

Analyze the loaded submissions.

```python
# Get statistics
validator = SubmissionValidator()
stats = validator.get_statistics(submissions_csv)

print("ðŸ“Š Submission Statistics")
print("=" * 50)
for key, value in stats.items():
    print(f"{key.replace('_', ' ').title()}: {value}")
```

## 6. Visualization: Code Length Distribution

```python
# Calculate code lengths
code_lengths = [len(sub['code']) for sub in submissions_csv]
submission_ids = [sub['submission_id'] for sub in submissions_csv]

# Create visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Bar chart
colors = sns.color_palette('viridis', len(submission_ids))
ax1.bar(submission_ids, code_lengths, color=colors)
ax1.set_xlabel('Submission ID', fontsize=12)
ax1.set_ylabel('Code Length (characters)', fontsize=12)
ax1.set_title('Code Length by Submission', fontsize=14, fontweight='bold')
ax1.grid(axis='y', alpha=0.3)

# Histogram
ax2.hist(code_lengths, bins=10, color='steelblue', edgecolor='black', alpha=0.7)
ax2.axvline(stats['avg_code_length'], color='red', linestyle='--', 
            linewidth=2, label=f"Mean: {stats['avg_code_length']:.0f}")
ax2.set_xlabel('Code Length (characters)', fontsize=12)
ax2.set_ylabel('Frequency', fontsize=12)
ax2.set_title('Code Length Distribution', fontsize=14, fontweight='bold')
ax2.legend()
ax2.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\nCode length range: {stats['min_code_length']} - {stats['max_code_length']} characters")
print(f"Average code length: {stats['avg_code_length']:.1f} characters")
```

## 7. Language Distribution

```python
# Count languages
from collections import Counter
languages = [sub['language'] for sub in submissions_csv]
lang_counts = Counter(languages)

# Pie chart
plt.figure(figsize=(8, 8))
colors = sns.color_palette('pastel')
plt.pie(lang_counts.values(), labels=lang_counts.keys(), autopct='%1.1f%%',
        startangle=90, colors=colors, textprops={'fontsize': 14})
plt.title('Programming Language Distribution', fontsize=16, fontweight='bold', pad=20)
plt.axis('equal')
plt.show()

print("Language breakdown:")
for lang, count in lang_counts.items():
    print(f"  {lang}: {count} submissions ({count/len(submissions_csv)*100:.1f}%)")
```
